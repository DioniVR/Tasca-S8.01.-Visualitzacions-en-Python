```python


# Importamos tabla transacciones y usuarios y las grabamos en dataframes.

query = "select * from transactions"
df_transactions = pd.read_sql_query(query,conexion)
query = "select * from companies"
df_companies = pd.read_sql_query(query,conexion)

df_transactions["Año"] = df_transactions["Timestamp"].dt.year

df_EJ5 = pd.merge(df_transactions,df_companies, left_on="business_id", right_on="company_id")

df_EJ5 = df_EJ5.sort_values(["Amount"])


plt.figure(figsize = (10,8))
ax1=sns.barplot(y="country", x ="Amount", data= df_EJ5, hue= "Año",  errorbar=None, estimator=np.sum)

plt.title("Importe ventas por pais de empresa y año",fontsize = 16)
plt.xlabel("Ventas",fontsize = 14,color ='black')
plt.ylabel("País de empresa",fontsize = 14,color ='black')
for container in ax1.containers:
    ax1.bar_label(container, fmt='%.0f', label_type='edge', color='black',fontsize=7)

plt.show()




```

En este gráfico he seleccionado  los países de las empresas, y los años como variables categóricas y he calculado el importe de las ventas.

En este gráfico se puede observas que las empresas de  Alemania, Suecia, UK y Canadá  son las que más venden respectivamente.
También se puede apreciar la importante diferencia  en las ventas entre 2021 y  2022. 

Estos datos no son concluyentes para concluir  si las ventas de 2022 son mayores que en 2021 o viceversa. Aunque no disponemos de datos para todos los meses de 2021 y  2022, para 2021 disponemos de datos de 10 meses mientras que para el  2022 sólo hay datos para el primer trimestre.

```python

# Importación de tablas y consulta :

query = "select * from products"
df_Eje_1_2 = pd.read_sql_query(query,conexion)  
query = "select * from transactions"
df_Eje1 = pd.read_sql_query(query,conexion) 
query = "select * from users"
df_users = pd.read_sql_query(query,conexion)
query = "select * from transaction_product"
df_transaction_product = pd.read_sql_query(query,conexion) 



df_users["birth_date"] = pd.to_datetime(df_users["birth_date"])
fecha_hoy = pd.Timestamp.now()
def age(birth_date):
    fecha_hoy = pd.Timestamp.now()
    return  fecha_hoy.year - birth_date.year
df_users["age"] = df_users["birth_date"].apply(age)
def age_range(age):
    if age <32:
        return "23-31"
    if age < 40:
        return  "32-39"
    else:
        return "40-45"
df_users["age_range"]  = df_users["age"].apply(age_range)


#merge dataframes

df_1= pd.merge(df_transactions,df_transaction_product ,left_on= "Id", right_on ="Transaction_id")
df_2 = pd.merge(df_1, df_products, left_on="product_id", right_on="Id")
df_3 = pd.merge(df_2, df_users, left_on="user_id", right_on="Id")
df_3.columns
df_Ej_1_7= pd.DataFrame(df_3, columns = ('Id','age','age_range','price','product_id'))
df_Ej_1_7 =df_Ej_1_7.groupby(["Id", "age", "age_range"])["product_id"].count().reset_index()
df_Ej_1_7 = df_Ej_1_7.rename(columns={"product_id": "Units purchased", "Id": "User_Id"})



# Graficamos



plt.figure(figsize=(10,10))
ax1 = sns.pairplot(df_Ej_1_7, hue="age_range", vars= ["age" ,"Units purchased"])
ax1.fig.suptitle('Análisis entre la edad usuarios y la cantidad de productos consumidos', y=1, fontsize = 10)
ax1.fig.tight_layout()
ax1.fig.subplots_adjust(top=0.90, right=0.8)
plt.show()

df_Ej_1_7.info()
```

```python
# importación de tablas y consulta.

query = "select * from transactions"
df_transactions = pd.read_sql_query(query,conexion)
query = "select * from users"
df_users = pd.read_sql_query(query,conexion)
query = "select * from products"
df_products = pd.read_sql_query(query,conexion)
query = "select * from transaction_product"
df_transaction_product = pd.read_sql_query(query,conexion)

df_analisis= pd.merge(df_transactions,df_users, left_on= "user_id", right_on ="Id")
df_analisis2 = pd.merge(df_analisis,df_transaction_product, left_on= "Id_x", right_on ="Transaction_id")
df_eje7= pd.merge(df_analisis2,df_products,left_on= "product_id", right_on ="Id")
df_eje7.columns
df_numeric = df_eje7[["Amount","weight","price","Lat", "Longitude"]]
df_numeric

# Graficar las correlaciones.

Correlation = round(df_numeric.corr(),3)
plt.figure(figsize=(10, 5))
sns.heatmap(data = Correlation, annot = True)
plt.suptitle("Correlación entre Amount, weight, price, Lat y Longitude ", fontsize=20, y=1)
plt.show()
```

Comentarios

Aquí observamos una matriz de corelaciones. Es una matriz simétrica.
En la diagonal tenemos las correlaciones de cada variable con ellos mismos con lo cual siempre es 1, es decir, correlación perfecta.

Las correlaciones encima de la diagonal es la misma tiene los mismos valores que la parte inferior.
Esto es así porque, por ejemplo, la correlacions de "Amount - Weight" es la misma que la correlación de "Weight-Amount". ESto pasa con todos los elementos de la matriz.

En general las correlacines son muy bajas todas por debajo de 0.03, indicando que no hay relación entre los datos, lo cual tiene sentido.

Por ejemplo, la longitud y la latitud son puntos en el espacio que de forma combinada nos da una posición en el espacio. Pero no tiene una relación de causalidad con las variables precio y weight.

La unica relación que podría tener sentido es el Amount con precio, pero hay que tener en cuenta que el amount es el importe de las transacciones y cada transacción puede tener varios productos con su respectivo precio. Aún así, la correlación es bajisima (0.011). Esto es debido a que el Amount es el amount de las transacciones y cada transacción está compuesta por diferentes productos que tienen diferentes precios.

Implementa un jointplot.



```python

# Importamos la tabla y consulta.

query = "select * from products"
df_ejer_2_2 = pd.read_sql_query(query,conexion) 
pd.set_option("display.max.columns", None)
df_ejer_2_2.head(10)

# Graficamos


sns.set_style('darkgrid')
plt.figure(figsize=(10, 5))
ax2 =sns.jointplot(x="weight",y= "price",data =df_ejer_2_2, kind="reg")
ax2.fig.suptitle('"Relación entre peso y precio"', y=1)
ax2.fig.tight_layout()
ax2.fig.subplots_adjust(top=0.95)
ax2.set_axis_labels("weight in kg","Price in $") 
plt.show()




```

Comentarios


El Joint plot nos permite integrar tres gráficos en uno. Por una parte tenemos en la parte superior, en el axis opuesto al peso, el histrograma de la variable peso.  En la parte lateral derecha, en el axis opuesto al precio, el histrograma de la variable Precio. Esto nos ayuda a ver como es la distribución de dichas variables. Finalmente, en el centro del gráfico, tenemos el scatterplot de las variables precio y peso.

En este caso, volvemos a ver que no hay ninguna relación entre dichas variables.

Adicionalmente, he añadido una linea de regresión que nos ayudará a interpretar los datos. Vemos que la linea de regresión entre las variables peso y precio es casi plana, lo cual indica que no hay ninguna relación entre ellas.

Implementa un violinplot combinat amb un altre tipus de gràfic.


```python
# importar 

query = "select * from products"
df_eje_3_1 = pd.read_sql_query(query,conexion) 


#Graficar
plt.figure(figsize=(10, 5))
ax3 = sns.JointGrid(data = df_eje_3_1, x= "weight",y= "price")
ax3.plot(sns.scatterplot, sns.violinplot)
ax3.fig.suptitle('"Relación entre peso y precio"', y=1)
ax3.fig.tight_layout()
ax3.fig.subplots_adjust(top=0.95)
ax3.set_axis_labels("weight in kg","Price in $") 
plt.show()
```

Comentarios

El gráfico de violin es buena manera de visualizar las variables numericas, ya que combina la el boxplot con la fuerza visual que nos da el diagrama de densidad.

En este caso lo he combinado con el jointplot porque de esta manera es como si tuvieramos 5 gráficos en un mismo gráfico; El diagrama de densidad de cada  variable, más el boxplot de cada variable y finalmente la relación entre las dos variables.

En el caso del precio, vemos que el precio medio está bastante en el centro de la distribución aunque más inclinado a la el quartil 1. Eso nos podría llevar a la conclusión de que la distribución es simétrica, sin embargo gracias al gráfico de densidad vemos que la distribución es bimodal. 

En el caso del peso, si observamos  en un boxplot, vemos que la distribución del peso es bastante simétrica, pero gracias al gráfico de densidad, vemos que la hay una aculumación de valores a la derecha (left-skewed)

En cuanto a la relación al precio y peso, aquí volvemos a ver que no hay ningúna correlación entre estas dos variables.

Exercici 2 - Nivell 3

Genera un FacetGrid per a visualitzar múltiples aspectes de les dades simultàniament.

```python
import mysql.connector
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


conexion = mysql.connector.connect(
  host="localhost",
  user="root",
  password="123456",
  database = "transacciones"
)

print(conexion)


# IMPORTACIÓN DE DATOS Y CONSULTA 

query = "select * from transactions"
df_transactions = pd.read_sql_query(query,conexion)
query = "select * from users"
df_users = pd.read_sql_query(query,conexion)
query = "select * from products"
df_products = pd.read_sql_query(query,conexion)
query = "select * from transaction_product"
df_transaction_product = pd.read_sql_query(query,conexion)

df_analisis= pd.merge(df_transactions,df_users, left_on= "user_id", right_on ="Id")
df_analisis2 = pd.merge(df_analisis,df_transaction_product, left_on= "Id_x", right_on ="Transaction_id")
df_eje_3_2= pd.merge(df_analisis2,df_products,left_on= "product_id", right_on ="Id") # Para sacar el year
df_eje_3_2["year"] = df_eje_3_2["Timestamp"].dt.year.astype("str")
New_df_eje_3_2 = df_eje_3_2[df_eje_3_2["product_name"].isin(["Direwolf Stannis","skywalker ewok","riverlands north"])]


# Gráfico

sns.set_style("white")
plt.figure(figsize = (10,5))
ax7= sns.FacetGrid(New_df_eje_3_2, col="country" )
ax7.map_dataframe(sns.countplot, y= "product_name", hue = "year",palette="dark:#1f77b4")
ax7.add_legend()
ax7.set_axis_labels("Units Sold","Product_name")
ax7.fig.suptitle('Productos más vendidos por país y año. ', y=1)
ax7.fig.tight_layout()
ax7.fig.subplots_adjust(top=0.8)
plt.show()


```

Comentarios

En este FacetGrid vemos los tres productos más veniddos,   en Canadá, United Stated y United kingdom  que son Direwolf Stannis ,skywalker ewok ,riverlands north, segregado por año 2021,2022.

Como sabemos, tanto los datos de 2021 como de 2022 están incompletos con lo cual no podemos decir que en  2022 se compró menos que en 2021.

Pero sí podemos ver el peso del mercado  Canadiense en la compra de productos de la empresa.  Cualquier shock econónimo que tenga lugar en Canadá afectará gravemente al número transacciones y ventas de la empresa.

Teniendo en cuenta esto, sería interesante promocionar los productos en Estados Unidos y en United Kingdom, no solo ayudará a diversificar el riesgo de una crisis en Canadá si no que  también, la venta de productos cuyo precio son los más caros hará que el volumen de ventas y la venta promedio crezcan.
